```python
import pandas
import seaborn
from matplotlib import pyplot
import numpy
```

```python
df = pandas.read_csv("nonp.csv")
```

```python
dims = (11.7, 8.27)
```

```python
fig,ax = pyplot.subplots(figsize=dims)
ax.set(yscale="log")
seaborn.lineplot(ax=ax, x="senders", y=" avgFwdDelay", data=df.query('senders%2 == 0'))
seaborn.scatterplot(ax=ax, x="senders", y=" avgFwdDelay", data=df.query('senders%2 == 0'))
```

```python
fig,ax = pyplot.subplots(figsize=dims)
ax.set(yscale="log")
seaborn.lineplot(ax=ax, x="senders", y=" throughput", data=df)
seaborn.scatterplot(ax=ax, x="senders", y=" throughput", data=df)
```

```python
fig,ax = pyplot.subplots(figsize=dims)
#ax.set(yscale="log")
seaborn.lineplot(ax=ax, x="senders", y=" efficiency", data=df)
seaborn.scatterplot(ax=ax, x="senders", y=" efficiency", data=df)
ax.grid(True)
```
