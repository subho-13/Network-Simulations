```python
allFiles = ["cdma.csv", "nonp.csv","onep.csv", "pp.csv"]
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dims = (10, 10)
```

```python
for file in ["onep.csv", "nonp.csv"]:
    df = pd.read_csv(file)
    df = df.sort_values(by=['senders', 'throughput', 'efficiency'])
    df = df.round(decimals=4)
    
    for column in ['efficiency', 'avgFwdDelay', 'throughput']:
        fig,ax = plt.subplots(figsize=dims)
        if column != "efficiency":
            ax.set(yscale="log")
        plot = sns.lineplot(ax=ax, x="senders", y=column, data=df)
        ax.grid(True)
        plot.figure.savefig(file+"_"+column+".png")
```

```python
df = pd.read_csv("pp.csv")
df = df.sort_values(by=['senders', 'throughput', 'efficiency'])
df = df.round(decimals=4)

palette = sns.color_palette("bright", len(df.senders.unique()))

for column in ['efficiency', 'avgFwdDelay', 'throughput']:
    fig,ax = plt.subplots(figsize=dims)
    if column != "efficiency":
        ax.set(yscale="log")
    plot = sns.lineplot(ax=ax, x="pvalue", y=column, hue='senders', palette=palette, data=df)
    ax.grid(True)
    plot.figure.savefig("pp_" + column+ ".png")
```

```python
df = pd.read_csv("cdma.csv")
df = df.sort_values(by=['senders', 'throughput'])
df = df.round(decimals=4)

for column in ['avgFwdDelay', 'throughput']:
    fig,ax = plt.subplots(figsize=dims)
    plot = sns.lineplot(ax=ax, x="senders", y=column, data=df)
    ax.grid(True)
    plot.figure.savefig("cdma_"+column+".png")
```
