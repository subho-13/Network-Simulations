```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

```python
dims = (20, 10)
```

```python
onep = pd.read_csv("onep.csv")
onep = onep.sort_values(by=['senders', 'throughput', 'efficiency'])
onep = onep.round(decimals=4)
```

```python
fig,ax = pyplot.subplots(figsize=dims)
ax.set(yscale="log")
sns.lineplot(ax=ax, x="senders", y="throughput", err_style="bars", data=onep)
ax.grid(True)
```
